## Bölüm 2 — Karşılaştırma Mantığı Nedir?

Hayatta sürekli karşılaştırma yaparız:

- *"Hava soğuk **mu**?"* → Evet ise mont giy.
- *"Sınavdan 50'nin üstünde **mi** aldım?"* → Evet ise geçtim.
- *"Yaşım 18'den büyük **mü**?"* → Evet ise ehliyet alabilirim.

Python da aynı mantıkla çalışır:  
Bir şeyi bir şeyle karşılaştırır ve sonuç olarak **iki cevaptan birini** verir:

| Cevap | Python'da | Anlamı |
|-------|-----------|--------|
| Evet  | `True`    | Doğru  |
| Hayır | `False`   | Yanlış |

````python
# Python'a bir soru soralım.
# 10, 5'ten büyük mü?
print(10 > 5)
````

````python
# 3, 7'den büyük mü?
print(3 > 7)
````

## Bölüm 3 — Karşılaştırma Operatörleri

Python'da 6 tane karşılaştırma operatörü var:

| Operatör | Anlamı | Örnek | Sonuç |
|----------|--------|-------|-------|
| `==` | Eşit mi? | `5 == 5` | `True` |
| `!=` | Eşit değil mi? | `5 != 3` | `True` |
| `>` | Büyük mü? | `7 > 3` | `True` |
| `<` | Küçük mü? | `2 < 8` | `True` |
| `>=` | Büyük veya eşit mi? | `5 >= 5` | `True` |
| `<=` | Küçük veya eşit mi? | `3 <= 1` | `False` |

**Dikkat:** `=` ile `==` farklıdır!  
- `=` → Değer **atama** (değişkene bir şey kaydetme)  
- `==` → **Karşılaştırma** (iki şey eşit mi diye sorma)

### Hepsini Tek Tek Deneyelim

````python
# == (Eşit Mi?)
print('5 == 5:', 5 == 5) # True
print('5 == 3:', 5 == 3) # False
````

````python
# != (Eşit Değil Mi?)
print('5 != 5:', 5 != 5) # False
print('5 != 3:', 5 != 3) # True
````

````python
# > ve < (Büyük mü? Küçük mü?)
print('10 > 3:', 10 > 3) # True
print('2 < 1:', 2 < 1) # False
````

````python
# >= ve <= (Büyük-eşit mi? Küçük-eşit mi?)
print('6 >= 5:', 5 >= 5) # True
print('5 >= 6:', 5 >= 6) # False
print('2 <= 3:', 3 <= 3) # True
print('3 <= 2:', 3 <= 2) # False
````

### Değişkenlerle Karşılaştırma

````python
puan = 75
gecme_notu = 50
print(f'Puan: {puan}, Geçme notu: {gecme_notu}')
print(f'Geçti mi? -> {puan >= gecme_notu}')
````

## Bölüm 4 — True ve False: Boolean Mantığı

`True` ve `False` Python'da özel değerlerdir.  
Bunlara **boolean** (mantıksal) değerler denir.

- Her karşılaştırma sonucu bir boolean'dır.
- Python'da `bool` adlı bir veri tipidir.
- Sadece iki değer alabilir: `True` veya `False`

**Dikkat:** Baş harfleri büyüktür! `True` doğru, `true` hata verir.

````python
# Boolean değerler
dogru = True
yanlis = False

print(dogru)
print(yanlis)
print(type(dogru))
print(type(yanlis))
````

````python
# Karşılaştırmaların sonucu boolean'dır
sonuc = 10 > 3
print(sonuc)
print(type(sonuc))
````

````python
# bool() ile değerleri boolean'a çevirebiliriz
print(f'bool(1): {bool(1)}') # True
print(f'bool(0): {bool(0)}') # False
print(f'bool(""): {bool("")}')
print(f'bool(" "): {bool(" ")}')
print(f'bool("a"): {bool("a")}')
````

## Bölüm 5 — `if` Nedir? (İlk Karar)

`if` İngilizce'de "eğer" demektir.

Yapısı çok basit:

```python
if koşul:
    # koşul True ise bu satır çalışır
```

- `koşul` → Bir karşılaştırma ifadesi (True veya False üretir)
- `:` → Satır sonunda iki nokta üst üste **zorunlu**
- Girinti → Alt satır **4 boşluk** içeriden başlar **zorunlu**

````python
# En basit if örneği
hava_sicakligi = 35
if hava_sicakligi > 30:
    print('Hava çok sıcak, bol bol su için.')
````

````python
# Koşul sağlanmazsa ne olur?
hava_sicakligi = 20
if hava_sicakligi > 30:
    print('Hava çok sıcak, bol bol su için.')
print('Program devam ediyor.')
````

## Bölüm 6 — Girintileme (Indentation) Neden Önemli?

Python'da girintileme (boşluklar) sadece güzellik için değildir — **zorunludur**.  
Python, hangi satırın `if`'e ait olduğunu **girintiden** anlar.

```
if koşul:
    bu satır if'in İÇİNDE (4 boşluk girinti)
    bu da if'in İÇİNDE
bu satır if'in DIŞINDA (girinti yok)
```

````python
# Doğru Girintileme
puan = 50
if puan >= 50:
    print('Tebrikler') # if'in İÇİNDE
    print('Sınavı geçtiniz') # if'in İÇİNDE
print('Program devam ediyor...') # if'in DIŞINDA - koşul sağlansa da sağlanmasa da (her zaman) çalışır.
````

````python
# Hatalı Girintileme

sayi = 10
if sayi > 5:
print('büyük')
````

## Bölüm 7 — `if - else` Yapısı

Bazen sadece "eğer doğruysa şunu yap" yetmez.  
"Doğru değilse de şunu yap" demek isteriz.

İşte `else` tam bunu yapar:

```python
if koşul:
    # koşul True ise burası çalışır
else:
    # koşul False ise burası çalışır
```

Düşün: bir yol ayrımı. Sola veya sağa gideceksin — ikisine birden gidemezsin.

````python
# if-else yaş kontrolü
yas = int(input('Lütfen yaşınızı giriniz:'))

if yas >= 18:
    print('Ehliyet alabilirsiniz.')
else:
    print(f'Ehliyet için {18 - yas} yıl daha beklemelisiniz.')
````

````python
# Çift mi? Tek mi?
sayi = int(input('Lütfen bir sayı giriniz:'))
if sayi % 2 == 0:
    print(f'{sayi} bir ÇİFT sayıdır')
else:
    print(f'{sayi} bir TEK sayıdır')
````

````python
# ALIŞTIRMA: Pozitif mi değil mi?
# Kullanıcıdan bir sayı al. Sayı pozitifse 'Pozitif' yazdır, değilse 'Negatif' yazdır.
sayi = int(input('Lütfen bir sayı giriniz:'))
if sayi < 0:
    print(f'{sayi} NEGATİF bir sayıdır.')
else:
    print(f'{sayi} POZİTİF bir sayıdır.')
````

## Bölüm 8 — `if - elif - else` Yapısı

Ya ikiden fazla seçenek varsa?

Mesela bir notu değerlendirirken sadece "geçti / kaldı" değil, farklı harf notları vermek istiyorsak?

İşte `elif` ("else if" kısaltması) bunu sağlar:

```python
if koşul_1:
    # koşul_1 True ise
elif koşul_2:
    # koşul_1 False ama koşul_2 True ise
elif koşul_3:
    # ilk ikisi False ama koşul_3 True ise
else:
    # hiçbiri True değilse
```

Python yukarıdan aşağıya kontrol eder. İlk `True` olan bloğu çalıştırır ve geri kalanları **atlar**.

````python
# else + if : elif
# Pozitif mi negatif mi sıfır mı kontrolü

sayi = int(input('Lütfen bir sayı giriniz:'))

if sayi < 0:
    print(f'{sayi} NEGATİF bir sayıdır.')
elif sayi == 0:
    print(f'{sayi} SIFIRDIR.')
else:
    print(f'{sayi} POZİTİF bir sayıdır.')
````

````python
# Not değerlendirme sistemi
puan = float(input('Sınav Notunu Gir (0-100)'))

if puan >= 90:
    print('Harf Notu: A')
elif puan >= 80:
    print('Harf Notu: B')
elif puan >= 70:
    print('Harf Notu: C')
elif puan >= 60:
    print('Harf Notu: D')
else:
    print('Harf Notu: F')
````

## Bölüm 12 — Mini Uygulama: Öğrenci Durum Raporu

Dersin son uygulaması! Bugün öğrendiğin her şeyi kullanacaksın.

**Senaryo:**  
Bir program yaz. Program şunları yapsın:

1. Kullanıcıdan **ad** alsın
2. Kullanıcıdan **yaş** alsın (sayıya çevir)
3. Kullanıcıdan **sınav notu** alsın (sayıya çevir)
4. Kişiye **adıyla** hitap etsin
5. Yaşına göre "Reşit" veya "Reşit değil" yazdırsın
6. Notuna göre harf notu ve durum mesajı versin:
   - 90+ → A → "Mükemmel!"
   - 80+ → B → "Çok iyi!"
   - 70+ → C → "İyi."
   - 60+ → D → "Geçer."
   - 60 altı → F → "Kaldın, tekrar çalışmalısın."
7. Sonuçları düzenli bir şekilde yazdırsın

### Örnek Çözüm

Kendin denedikten sonra aşağıdaki çözüme bak:

````python
# ========================================
# MİNİ UYGULAMA: Öğrenci Durum Raporu
# Örnek Çözüm
# ========================================

# 1. Bilgileri al
ad = input("Adınız: ")
yas = int(input("Yaşınız: "))
not_puan = int(input("Sınav notunuz (0-100): "))

# 2. Yaş kontrolü
if yas >= 18:
    yas_durumu = "Reşit"
else:
    yas_durumu = "Reşit değil"

# 3. Not değerlendirme
if not_puan >= 90:
    harf = "A"
    mesaj = "Mükemmel!"
elif not_puan >= 80:
    harf = "B"
    mesaj = "Çok iyi!"
elif not_puan >= 70:
    harf = "C"
    mesaj = "İyi."
elif not_puan >= 60:
    harf = "D"
    mesaj = "Geçer."
else:
    harf = "F"
    mesaj = "Kaldın, tekrar çalışmalısın."

# 4. Sonuçları yazdır
print("\n" + "=" * 35)
print("     ÖĞRENCİ DURUM RAPORU")
print("=" * 35)
print(f"Ad       : {ad}")
print(f"Yaş      : {yas} ({yas_durumu})")
print(f"Not      : {not_puan}")
print(f"Harf Notu: {harf}")
print("-" * 35)
print(f"Sonuç    : {mesaj}")
print("=" * 35)
````
