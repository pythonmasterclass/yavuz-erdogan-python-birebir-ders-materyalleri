## Bölüm 2 — `min()` ve `max()`: En Küçük ve En Büyük

Bu iki fonksiyon bir listedeki **en küçük** ve **en büyük** değeri bulur.

```python
min(liste)  # En küçük değer
max(liste)  # En büyük değer
```

Hem sayılarla hem de stringlerle çalışır.

````python
# Sayılarla min() ve max() kullanımı
notlar = [78, 92, 55, 100, 67, 43, 88]

print('Notlar:',  notlar)
print('En yüksek not:',  max(notlar))
print('En düşük not:',  min(notlar))
````

````python
# Stringlerle min() ve max() kullanımı
isimler = ['Zeynep', 'Ali', 'Mehmet', 'Ayşe']
print('İsimler:', isimler)
print('Alfabetik ilk:', min(isimler))
print('Alfabetik son:', max(isimler))
````

````python
# min/max ile indeks bulma
ogrenciler = ['Ali', 'Ayşe', 'Mehmet', 'Zeynep', 'Can']
notlar = [78, 92, 55, 100, 67]

en_yuksek_not = max(notlar)
en_yuksek_notu_alan_ogrenci_indeksi = notlar.index(en_yuksek_not)
en_yuksek_notu_alan_ogrenci = ogrenciler[en_yuksek_notu_alan_ogrenci_indeksi]

print(f"En yüksek not: {en_yuksek_not} -> {en_yuksek_notu_alan_ogrenci}")

en_dusuk_not = min(notlar)
en_dusuk_notu_alan_ogrenci_indeksi = notlar.index(en_dusuk_not)
en_dusuk_notu_alan_ogrenci = ogrenciler[en_dusuk_notu_alan_ogrenci_indeksi]

print(f"En düşük not: {en_dusuk_not} -> {en_dusuk_notu_alan_ogrenci}")
````

## Bölüm 3 — `sum()`: Toplama Fonksiyonu

`sum()` bir sayı listesindeki tüm elemanları toplar.

```python
sum(liste)  # Tüm elemanların toplamı
```

**Dikkat:** `sum()` sadece sayı listeleriyle çalışır. String listesi verirsen hata alırsın.

````python
# Summation'dan geliyor.
# Basit toplama
notlar = [78, 92, 55, 100, 67]
toplam = sum(notlar)
print(f"Notlar Listesi: {notlar}")
print(f"Notlar Listesinin Toplamı: {toplam}")
````

````python
# Harcamalar Toplamı
harcamalar = [45.50, 33.75, 19.99, 120.00, 99.99, 73.74]
toplam = sum(harcamalar)
print(f"Harcamalar Listesi: {harcamalar}")
print(f"Harcamalar Listesinin Toplamı: {toplam:.2f} TL")

# ÖDEV: Bir liste oluştur. İçerisinde 10 eleman olsun. 50'nin üzerindekileri sadece topla. Dilimleme ve if-elif-else kullanabilirsin.
````

## Bölüm 4 — Ortalama Hesaplama: `sum()` + `len()` Birlikte

Python'da hazır bir `average()` fonksiyonu yok.  
Ama ortalamayı kolayca hesaplayabilirsin:

```python
ortalama = sum(liste) / len(liste)
```

Bu kadar! Toplam ÷ eleman sayısı = ortalama.

````python
notlar = [78, 92, 55, 100, 63, 67, 88, 73, 40]
toplam = sum(notlar)
adet = len(notlar)
ortalama = toplam / adet
print(f"Notlar: {notlar}")
print(f"Toplam: {toplam}")
print(f"Adet: {adet}")
print(f"Ortalama: {ortalama:.2f}")
````

````python
# Aylık harcama ortalaması
aylik_harcamalar = [25000, 31000, 28000, 42000, 19000, 35000]
aylar = ['Ocak', 'Şubat', 'Mart', 'Nisan', 'Mayıs', 'Haziran']

ortalama = sum(aylik_harcamalar) / len(aylik_harcamalar)
en_pahali_ay = aylar[aylik_harcamalar.index(max(aylik_harcamalar))]
en_ucuz_ay = aylar[aylik_harcamalar.index(min(aylik_harcamalar))]

print(f"Toplam Harcama: {sum(aylik_harcamalar)} TL")
print(f"Aylık Ortalama Harcama: {ortalama} TL")
print(f"En Pahalı Ay: {en_pahali_ay} ve {max(aylik_harcamalar)} TL")
print(f"En Ucuz Ay: {en_ucuz_ay} ve {min(aylik_harcamalar)} TL")
````

## Bölüm 5 — `sorted()` vs `.sort()`: Önemli Fark

İkisi de sıralama yapar ama aralarında **kritik bir fark** var:

| Özellik | `.sort()` | `sorted()` |
|---------|-----------|------------|
| Kullanım | `liste.sort()` | `sorted(liste)` |
| Orijinal listeyi değiştirir mi? | **Evet** | **Hayır** |
| Geri dönüş | `None` | **Yeni bir liste** |
| Ne zaman kullan? | Orijinali değiştirmek istersen | Orijinali korumak istersen |

````python
sayilar_a = [2, 6, 9, 7, 3, 1]
print(sayilar_a)
print(id(sayilar_a))
sayilar_a.sort()
print(sayilar_a)
print(id(sayilar_a))
sayilar_b = [2, 6, 9, 7, 3, 1]
print(sayilar_b)
print(id(sayilar_b))
sirali_sayilar = sorted(sayilar_b)
print(sirali_sayilar)
print(id(sirali_sayilar))
print(sayilar_b)
print(id(sayilar_b))
````

````python
sayilar = [2, 6, 9, 7, 3, 1]
kucukten_buyuge = sorted(sayilar)
buyukten_kucuge = sorted(sayilar, reverse=True)

print(kucukten_buyuge)
print(buyukten_kucuge)
````

## Bölüm 6 — Liste Birleştirme ve Kopyalama

Listelerle yapılabilecek birkaç önemli işlem daha var.

### 6.1 — Liste Birleştirme: `+` Operatörü

````python
# İki listeyi birleştirme
meyveler = ['elma', 'armut', 'muz']
sebzeler = ['domates', 'biber', 'patlıcan']

manav = meyveler + sebzeler
print(f"Meyveler reyonundaki ürünler: {meyveler}")
print(f"Sebzeler reyonundaki ürünler: {sebzeler}")
print(f"Manav bölümündeki ürünler: {manav}")
````

### 6.2 — `.extend()` Metodu

````python
# .extend() bşr listeye başka bir listenin elemanlarını ekler
liste1 = [1, 2, 3]
liste1.append([4, 5]) # [4, 5]'i TEK eleman olarka ekleyecek.
print(liste1)

liste2 = [1, 2, 3]
liste2.extend([4, 5]) # 4 ve 5'i AYRI AYRI ekleyecek.
print(liste2)
````

````python
liste_a = [1, 2, 3]
liste_b = [4, 5, 6]
liste_a.extend(liste_b)
print(liste_a)
print(liste_b)
````

### 6.3 — Liste Kopyalama

````python
orijinal = [1, 2, 3]
kopya_degil = orijinal

kopya_degil.append(4)
print(kopya_degil)
print(orijinal)
````

````python
orijinal = [1, 2, 3]

# yol 1: .copy() metodu
kopya1 = orijinal.copy()

# yol 2: dilimleme
kopya2 = orijinal[:]

# yol 3: list() fonksiyonu kullanılır
kopya3 = list(orijinal)

# Şimdi kopyaları değiştirelim
kopya1.append(4)
kopya2.append(5)
kopya3.append(6)

print(orijinal)
print(kopya1)
print(kopya2)
print(kopya3)
````

### 6.4 — Liste Tekrarlama: `*` Operatörü

````python
sifirlar = [0] * 5
print(sifirlar)

desen = [1, 2, 3] * 3
print(desen)
````

## Bölüm 7 — Tüm Liste Fonksiyonları ve Metotları

### Hazır Fonksiyonlar (Built-in)

| Fonksiyon | Ne Yapar | Örnek | Sonuç |
|-----------|----------|-------|-------|
| `len(liste)` | Eleman sayısı | `len([1,2,3])` | `3` |
| `min(liste)` | En küçük değer | `min([5,2,8])` | `2` |
| `max(liste)` | En büyük değer | `max([5,2,8])` | `8` |
| `sum(liste)` | Tüm elemanların toplamı | `sum([1,2,3])` | `6` |
| `sorted(liste)` | Sıralı kopya döndürür | `sorted([3,1,2])` | `[1,2,3]` |

### Liste Metotları

| Metot | Ne Yapar | Listeyi Değiştirir? |
|-------|----------|--------------------|
| `.append(x)` | Sona ekler | Evet |
| `.insert(i, x)` | Araya ekler | Evet |
| `.extend(liste)` | Başka listenin elemanlarını ekler | Evet |
| `.remove(x)` | Değerle siler | Evet |
| `.pop(i)` | İndeksle çıkarır | Evet |
| `.sort()` | Yerinde sıralar | Evet |
| `.reverse()` | Yerinde ters çevirir | Evet |
| `.copy()` | Gerçek kopya oluşturur | Hayır (yeni liste) |
| `.count(x)` | Kaç kez geçiyor | Hayır |
| `.index(x)` | İlk indeksini bulur | Hayır |
| `.clear()` | Tümünü siler | Evet |

### Operatörler

| Operatör | Ne Yapar | Örnek |
|----------|----------|-------|
| `+` | İki listeyi birleştirir | `[1,2] + [3,4]` → `[1,2,3,4]` |
| `*` | Listeyi tekrarlar | `[0] * 3` → `[0,0,0]` |
| `in` | Eleman var mı? | `3 in [1,2,3]` → `True` |

## Bölüm 8 — Uygulama 1: Öğrenci Not Analiz Sistemi

Bir sınıftaki öğrencilerin sınav notlarını analiz eden bir program.

````python
# ========================================
# UYGULAMA 1: Öğrenci Not Analiz Sistemi
# ========================================

ogrenciler = ["Ali", "Ayşe", "Mehmet", "Zeynep", "Can", "Elif", "Burak"]
notlar = [78, 92, 45, 88, 67, 95, 53]

# Temel İstatistikler
print('=' * 40)
print('ÖĞRENCİ NOT ANALİZ RAPORU')
print('=' * 40)

print(f"Öğrenci Sayısı : {len(ogrenciler)}")
print(f"Sınavdan Alınan Toplam Puan : {sum(notlar)}")
print(f"Sınıf Ortalaması : {sum(notlar) / len(notlar):.2f}")
print(f"En Yüksek Not : {max(notlar)}")
print(f"En Düşük Not : {min(notlar)}")
print(f"Not Aralığı : {max(notlar) - min(notlar)}")

# En başarılı ve başarısız öğrenci
print('-' * 40)
en_yuksek_i = notlar.index(max(notlar))
en_dusuk_i = notlar.index(min(notlar))

print(f"Birinci : {ogrenciler[en_yuksek_i]} ({notlar[en_yuksek_i]})")
print(f"Sonuncu : {ogrenciler[en_dusuk_i]} ({notlar[en_dusuk_i]})")

# Sıralı liste
print('-' * 40)
print(f"Notlar (büyükten, küçüğe): {sorted(notlar, reverse=True)}")
print('=' * 40)
````
