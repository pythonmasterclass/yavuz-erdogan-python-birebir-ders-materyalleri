# **Yavuz Erdoğan** - Ders 1 : Uygulama Dosyası

Bu derste amacımız Python'ı sıfırdan öğrenmek değil, temel parçaları yerine oturtmak.

Bugün özellikle şunlara odaklanacağız:

- `print()` ile çıktı üretme
- değişken mantığı
- temel veri türlerini fark etme
- `input()` ile kullanıcıdan veri alma
- tip dönüşümü
- basit matematiksel işlemler
- küçük bir mini uygulama yapma

Bu dersin sonunda hedefimiz şudur:

**"Kodun mantığını daha net görüyorum ve temel şeyleri kendi başıma yapabiliyorum."**

**Not:** Hata yapmak bu dersin normal bir parçası. Hata görmek, yanlış yolda olduğun anlamına gelmez. Genelde sadece Python'ın senden biraz daha net olmanı istemesi anlamına gelir.

## 1. **`print()`** ile hızlı başlangıç

Python'da çoğu zaman ilk yaptığımız şey ekrana bir şey yazdırmaktır.

Bunun için `print()` kullanırız.

`print()`, parantez içindeki değeri ekrana gösterir.

## 2. **Değişkenler**

Değişkeni, bir bilgiyi geçici olarak sakladığımız kutu gibi düşünebiliriz.

Örneğin bir isim, yaş veya şehir bilgisini değişkende tutabiliriz.

Bu sayede aynı veriyi tekrar tekrar yazmak yerine bir isim verip gerektiğinde kullanırız.

## 3. Veri türü mantığına giriş

Python, her veriyi aynı şekilde görmez.

Örneğin:

- `"25"` bir metindir *(string)*
- `25` bir sayıdır *(integer)*

İkisi dışarıdan benzer görünebilir ama Python için aynı şey değildir.

Bunu anlamak çok önemlidir çünkü ileride bazı hataların sebebi tam olarak bu fark olacak.

## 4. **`input()`** ile kullanıcıdan veri alma

Şimdi biraz daha etkileşimli bir noktaya geçiyoruz.

`input()` sayesinde kullanıcıdan bilgi alabiliriz.

Kullanıcının yazdığı bilgi bir değişkende tutulur ve sonra kullanılabilir.

## 5. input neden gerektiğinde sayı gibi davranmaz?

Burada öğrencilerin çok sık takıldığı bir nokta var.

Kullanıcı klavyeden `25` yazsa bile, `input()` bunu doğrudan sayı olarak değil, metin olarak alır.

Yani kullanıcı 25 yazsa bile Python bunu çoğu zaman `"25"` gibi görür.

Şimdi bunu görelim.

## 6. Tip dönüşümü

Eğer kullanıcıdan aldığımız veriyle matematik yapmak istiyorsak, o veriyi uygun türe çevirmemiz gerekir.

Sık kullanılan iki dönüşüm:

- `int()` → tam sayıya çevirir
- `float()` → ondalıklı sayıya çevirir

Bu sayede Python o veriyi artık sayı gibi işlemeye başlar.

## 7. Aritmetik operatörler

Python ile temel matematik işlemleri yapabiliriz.

En sık kullanılan operatörler:

- `+` toplama
- `-` çıkarma
- `*` çarpma
- `/` bölme
- `%` kalan bulma
- `**` üs alma

## 8. Küçük birleştirme alıştırmaları

Şimdiye kadar gördüğümüz parçaları bir araya getiriyoruz:

- veri al
- değişkende tut
- gerekiyorsa dönüştür
- işlem yap
- sonucu anlamlı bir şekilde göster

Aslında programlama çoğu zaman tam olarak budur.

## 9. Temel hata mantığı

Kod yazarken hata görmek çok normaldir.

Önemli olan hiç hata yapmamak değil, hatanın neden çıktığını yavaş yavaş anlamaya başlamaktır.

Başlangıçta en sık görülen hatalar şunlardır:

- tırnağı kapatmamak
- parantezi eksik bırakmak
- yanlış girintileme
- sayı beklenen yerde metin kullanmak

Hata, başarısızlık işareti değildir.
Hata, Python'ın "burayı tekrar kontrol et" deme biçimidir.

## 10. Ders sonu mini uygulama

Şimdi bugünkü parçaları tek bir küçük uygulamada birleştireceğiz.

Bu uygulama:

- kullanıcıdan ad alacak
- yaş alacak
- iki sayı alacak
- yaş bilgisini kullanarak kişisel bir mesaj verecek
- iki sayı üzerinde basit işlem yapacak
- sonucu düzenli şekilde ekrana yazdıracak

Bu küçük uygulama, bugünkü dersin özeti gibi düşünülebilir.
