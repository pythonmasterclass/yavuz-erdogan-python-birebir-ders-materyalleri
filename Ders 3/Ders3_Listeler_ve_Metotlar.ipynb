## Bölüm 2 — Liste Nedir?

Liste, birden fazla değeri **sıralı** şekilde tutan bir veri yapısıdır.

Gerçek hayattan analoji:  
Bir **alışveriş listesi** düşün — her maddede bir ürün yazıyor ve sıralılar.

Python'da liste oluşturmak için **köşeli parantez** `[ ]` kullanırız:

```python
liste_adi = [eleman1, eleman2, eleman3]
```

**Önemli özellikler:**
- Sıralıdır (ilk eleman, ikinci eleman... diye gider)
- Değiştirilebilir (sonradan eleman ekleyip çıkarabilirsin)
- Farklı veri tiplerini bir arada tutabilir
- `list` adlı bir veri tipidir

````python
# Farklı türde listeler
meyveler = ['elma', 'armut', 'muz', 'çilek']
notlar = [85, 92, 65, 50, 34]
karisik = ['Yavuz', 22, True, 1.78]
bos_liste = []

print('Meyveler:', meyveler)
print('Notlar:', notlar)
print('Karışık:', karisik)
print('Boş Liste:', bos_liste)
````

````python
# Listenin tipi
print(type(meyveler))
print(type(notlar))
````

````python
print(type([1, 2, 3]))

listem = ['Ahmet', 23, 1.92, True, False, ['Mehmet', 35]]
print(listem)
````

## Bölüm 3 — Elemanlara Erişim: İndeksleme

Listedeki her elemanın bir **sıra numarası** (indeks) vardır.  
**Çok önemli:** Python'da sayma **0'dan** başlar!

```
meyveler = ["elma", "armut", "muz", "çilek"]
              0        1       2       3
```

Bir elemana erişmek için: `liste[indeks]`

````python
meyveler = ['elma', 'armut', 'muz', 'çilek']
print(meyveler[0]) # İlk eleman
print(meyveler[1]) # İkinci eleman
print(meyveler[2]) # Üçüncü eleman
print(meyveler[3]) # Dördüncü eleman
````

````python
# meyveler[4] yazarsam ne olur?
print(meyveler[4]) # IndexError
````

````python
meyveler = ['elma', 'armut', 'muz', 'çilek']
print(meyveler[2])
````

## Bölüm 4 — Negatif İndeksleme: Sondan Erişim

Python'da listenin **sonundan** da erişebilirsin:

```
meyveler = ["elma", "armut", "muz", "çilek"]
              0        1       2       3      → pozitif indeks
             -4       -3      -2      -1      → negatif indeks
```

- `-1` → Son eleman
- `-2` → Sondan ikinci
- `-3` → Sondan üçüncü
- ...

````python
meyveler = ['elma', 'armut', 'muz', 'çilek']
print('Son eleman', meyveler[-1])
print('Sondan ikinci', meyveler[-2])
````

## Bölüm 5 — `len()` Fonksiyonu: Listenin Uzunluğu

`len()` fonksiyonu bir listenin kaç elemana sahip olduğunu söyler.

Bu çok işe yarar çünkü:
- Listenin boyutunu bilmeden son elemana erişebilirsin
- Listeyle ilgili kararlar verebilirsin ("liste boş mu?" gibi)

````python
meyveler = ['elma', 'armut', 'muz', 'çilek']
notlar = [85, 92, 65, 50, 34]
bos = []

print('Meyveler:', len(meyveler), 'eleman')
print('Notlar:', len(notlar), 'eleman')
print('Boş:', len(bos), 'eleman')
````

````python
# len() ile son elemana erişim
meyveler = ['elma', 'armut', 'muz', 'çilek']
son_indeks = len(meyveler) - 1
print(f'Son indeks: {son_indeks}')
print(f'Son eleman: {meyveler[son_indeks]}')
````

## Bölüm 6 — Dilimleme (Slicing): Listenin Bir Parçasını Almak

Bazen listenin tamamını değil, sadece **bir bölümünü** almak isteriz.

Söz dizimi: `liste[başlangıç:bitiş]`

**Önemli kurallar:**
- `başlangıç` indeksi **dahil**
- `bitiş` indeksi **dahil değil** (bir öncesine kadar alır)
- Başlangıç yazmazsan → baştan başlar
- Bitiş yazmazsan → sona kadar gider

## Bölüm 8 — En Sık Kullanılan Liste Metotları

Metotlar, listelerin üzerinde işlem yapan **hazır fonksiyonlardır**.  
Kullanımı: `liste.metot()`

Bugün göreceğimiz metotlar:

| Metot | Ne Yapar | Örnek |
|-------|----------|-------|
| `.append()` | Sona eleman ekler | `liste.append("yeni")` |
| `.insert()` | Belirli konuma ekler | `liste.insert(1, "yeni")` |
| `.remove()` | Değere göre siler | `liste.remove("eski")` |
| `.pop()` | İndekse göre çıkarır | `liste.pop(2)` |
| `.sort()` | Sıralar | `liste.sort()` |
| `.reverse()` | Ters çevirir | `liste.reverse()` |
| `.count()` | Kaç kez geçiyor sayar | `liste.count("elma")` |
| `.index()` | İlk indeksini bulur | `liste.index("elma")` |
| `.clear()` | Tüm elemanları siler | `liste.clear()` |

Her birini teker teker görelim.

### 8.1 — `.append()` → Sona Eleman Ekle

### 8.2 — `.insert()` → Belirli Konuma Ekle

### 8.3 — `.remove()` → Değere Göre Sil

### 8.4 — `.pop()` → İndekse Göre Çıkar (ve Geri Ver)

### 8.5 — `.sort()` → Sırala

### 8.6 — `.reverse()` → Ters Çevir

### 8.7 — `.count()` → Kaç Kez Geçiyor?

### 8.8 — `.index()` → İlk Bulunduğu İndeks

### 8.9 — `.clear()` → Tüm Elemanları Sil

## Bölüm 9 — `in` Operatörü: Listede Var mı?

Bir elemanın listede olup olmadığını kontrol etmek için `in` kullanırız.  
Sonuç bir **boolean** döner: `True` veya `False`.


Bu, `if` ile birlikte çok güçlü bir araç olur!

## Bölüm 10 — Özet Tablo: Metotların Karşılaştırması

| İşlem | Metot | Parametre | Listeyi Değiştirir mi? | Geri Dönüş |
|-------|-------|-----------|----------------------|------------|
| Sona ekle | `.append(x)` | Eklenecek değer | Evet | `None` |
| Araya ekle | `.insert(i, x)` | İndeks, Değer | Evet | `None` |
| Değerle sil | `.remove(x)` | Silinecek değer | Evet | `None` |
| İndeksle çıkar | `.pop(i)` | İndeks (opsiyonel) | Evet | Çıkarılan eleman |
| Sırala | `.sort()` | `reverse=True` (opsiyonel) | Evet | `None` |
| Ters çevir | `.reverse()` | — | Evet | `None` |
| Say | `.count(x)` | Aranacak değer | Hayır | Sayı |
| İndeks bul | `.index(x)` | Aranacak değer | Hayır | İndeks |
| Hepsini sil | `.clear()` | — | Evet | `None` |
| Uzunluk | `len(liste)` | — | Hayır | Sayı |
| Var mı? | `x in liste` | — | Hayır | `True` / `False` |

### Örnek Çözüm

Kendin denedikten sonra aşağıdaki çözüme bak:

````python
# ========================================
# MİNİ UYGULAMA: Sınıf Not Defteri
# Örnek Çözüm
# ========================================

# 1. Listeleri oluştur
ogrenciler = ["Ali", "Ayşe", "Mehmet", "Zeynep", "Can"]
notlar = [85, 42, 78, 95, 60]

# 2. Tüm öğrencileri ve notlarını yazdır
print("=" * 30)
print("    SINIF NOT DEFTERİ")
print("=" * 30)
print(f"{ogrenciler[0]:10s} → {notlar[0]}")
print(f"{ogrenciler[1]:10s} → {notlar[1]}")
print(f"{ogrenciler[2]:10s} → {notlar[2]}")
print(f"{ogrenciler[3]:10s} → {notlar[3]}")
print(f"{ogrenciler[4]:10s} → {notlar[4]}")
print("-" * 30)

# 3. En yüksek not
en_yuksek = max(notlar)
en_yuksek_indeks = notlar.index(en_yuksek)
print(f"En yüksek not: {en_yuksek} ({ogrenciler[en_yuksek_indeks]})")

# 4. En düşük not
en_dusuk = min(notlar)
en_dusuk_indeks = notlar.index(en_dusuk)
print(f"En düşük not : {en_dusuk} ({ogrenciler[en_dusuk_indeks]})")

# 5. Kullanıcıdan öğrenci arama
print("\n" + "-" * 30)
aranan = input("Bir öğrenci adı gir: ")

if aranan in ogrenciler:
    indeks = ogrenciler.index(aranan)
    print(f"{aranan}'ın notu: {notlar[indeks]}")
    
    # Not kontrolü
    if notlar[indeks] >= 50:
        print("Durum: Geçti")
    else:
        print("Durum: Kaldı")
else:
    print(f"{aranan} bu sınıfta yok.")

print("=" * 30)
````
