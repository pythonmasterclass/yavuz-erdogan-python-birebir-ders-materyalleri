### Soru 16 — Koşul Çıktı Tahmini

Aşağıdaki kodun çıktısını **önce kafandan tahmin et**, sonra çalıştırarak kontrol et.

```python
x = 15

if x > 20:
    print("A")
elif x > 10:
    print("B")
elif x > 5:
    print("C")
else:
    print("D")

print("Bitti")
```

**Sorular:**
1. Hangi harf yazdırılır?
2. `x > 10` ve `x > 5` ikisi de `True` olduğu halde neden sadece `"B"` yazdırılır?
3. `"Bitti"` her durumda yazdırılır mı? Neden?

````python
# Soru 16 — Tahminini yaz, sonra çalıştır:
# Tahminim: 

x = 15

if x > 20:
    print("A")
elif x > 10:
    print("B")
elif x > 5:
    print("C")
else:
    print("D")

print("Bitti")
````

<details>
<summary>💡 Açıklama</summary>

Çıktı: `B` ve `Bitti`. Python `elif` zincirinde ilk `True` olan koşulu çalıştırır ve geri kalanını **atlar**. `"Bitti"` if bloğunun dışında olduğu için (girintisiz) her zaman çalışır.
</details>
