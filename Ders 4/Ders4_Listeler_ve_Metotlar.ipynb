## Bölüm 2 — Liste Nedir?

Liste, birden fazla değeri **sıralı** şekilde tutan bir veri yapısıdır.

meyveler = ["elma", "armut", "muz", "çilek"]

## Bölüm 6 — Dilimleme (Slicing): Listenin Bir Parçasını Almak

Bazen listenin tamamını değil, sadece **bir bölümünü** almak isteriz.

Söz dizimi: `liste[başlangıç:bitiş]`

**Önemli kurallar:**
- `başlangıç` indeksi **dahil**
- `bitiş` indeksi **dahil değil** (bir öncesine kadar alır)
- Başlangıç yazmazsan → baştan başlar
- Bitiş yazmazsan → sona kadar gider

````python
sayilar = [10, 20, 30, 40, 50, 60, 70]
#          0   1   2   3   4   5   6

print(f"Tüm Liste: {sayilar}")
print(f"1.indexten 4.indexe kadar olan sayılar: {sayilar[1:4]}") # 20, 30, 40
print(f"(0:3) Baştan 3.indexe kadar olan sayılar: {sayilar[0:3]}") # 10, 20, 30
print(f"(:3) Baştan 3.indexe kadar olan sayılar: {sayilar[:3]}") # 10, 20, 30
print(f"2.indexten 6.indexe kadar olan sayılar: {sayilar[2:6]}") # 30, 40, 50, 60
````

````python
# Kısa yazımlar
print(f"Başlangıçtan 3.indexe kadar olan sayılar: {sayilar[:3]}") # 10, 20, 30
print(f"4.indexten sona kadar olan sayılar: {sayilar[4:]}") # 50, 60, 70
print(f"Listenin kopyası: {sayilar[:]}") # Tüm listenin kopyasını alır
````

````python
harfler = ['a', 'b', 'c']
sahte_kopya = harfler
# 2 listenin de RAM'deki adresini alalım
print("harfler listesinin RAM'deki adresi", id(harfler))
print("sahte_kopya listesinin RAM'deki adresi", id(sahte_kopya))
# sahte_kopya üzerinde bir değişiklik yapalım
sahte_kopya[0] = 'e'
print("harfler listesinin değişiklik sonrası hâli", harfler)
print("sahte_kopya listesinin değişiklik sonrası hâli", sahte_kopya)
````

````python
sayilar = [10, 20, 30]
# Tüm listenin gerçek bir kopyasını oluşturalım
gercek_kopya = sayilar[:]
# 2 listenin de RAM'deki adresini alalım
print("sayilar listesinin RAM'deki adresi", id(sayilar))
print("gercek_kopya listesinin RAM'deki adresi", id(gercek_kopya))
# Kopyada bir değişiklik yapalım
gercek_kopya[0] = 0
gercek_kopya[1] = 10
gercek_kopya[2] = 10

print("sayilar listesinin değişiklik sonrası hâli", sayilar)
print("gercek_kopya listesinin değişiklik sonrası hâli", gercek_kopya)
````

````python
# Negatif indeksle dilimleme
sayilar = [10, 20, 30, 40, 50, 60, 70]
#          0   1   2   3   4   5   6
#          -7  -6  -5  -4  -3  -2  -1
print(f"Son 3 Eleman: {sayilar[-3:]}")
print(f"Son 2 Eleman Hariç Tüm Liste: {sayilar[:-2]}")
print(f"70, 60 yazdırma: {sayilar[-1:-3:-1]}")
````

## Bölüm 8 — En Sık Kullanılan Liste Metotları

Metotlar, listelerin üzerinde işlem yapan **hazır fonksiyonlardır**.  
Kullanımı: `liste.metot()`

Bugün göreceğimiz metotlar:

| Metot | Ne Yapar | Örnek |
|-------|----------|-------|
| `.append()` | Sona eleman ekler | `liste.append("yeni")` |
| `.insert()` | Belirli konuma ekler | `liste.insert(1, "yeni")` |
| `.remove()` | Değere göre siler | `liste.remove("eski")` |
| `.pop()` | İndekse göre çıkarır | `liste.pop(2)` |
| `.sort()` | Sıralar | `liste.sort()` |
| `.reverse()` | Ters çevirir | `liste.reverse()` |
| `.count()` | Kaç kez geçiyor sayar | `liste.count("elma")` |
| `.index()` | İlk indeksini bulur | `liste.index("elma")` |
| `.clear()` | Tüm elemanları siler | `liste.clear()` |

Her birini teker teker görelim.

### 8.1 — `.append()` → Sona Eleman Ekle

````python
# HER ZAMAN LİSTENİN SONUNA EKLEME YAPAR
meyveler = ['elma', 'armut']
print(f'.append("muz") öncesi meyveler listesi: {meyveler}')
meyveler.append("muz")
print(f'.append("muz") sonrası meyveler listesi: {meyveler}')

print(f'.append("çilek") öncesi meyveler listesi: {meyveler}')
meyveler.append("çilek")
print(f'.append("çilek") sonrası meyveler listesi: {meyveler}')
````

### 8.2 — `.insert()` → Belirli Konuma Ekle

````python
sayilar = [10, 20, 40]
sayilar.insert(2, 30)
print(sayilar)
````

### 8.3 — `.remove()` → Değere Göre Sil

````python
meyveler = ['elma', 'armut', 'muz', 'çilek', 'elma']
meyveler.remove('elma')
print(meyveler)
````

### 8.4 — `.pop()` → İndekse Göre Çıkar (ve Geri Ver)

````python
meyveler = ['elma', 'armut', 'muz', 'çilek']
son = meyveler.pop()
print('Son eleman :', son)
print('Kalan meyveler :', meyveler)
````

````python
meyveler = ['elma', 'armut', 'muz', 'çilek']
# 2.indeksteki elemanı çıkar
cikarilan = meyveler.pop(2)
print(f'Çıkarılan: {cikarilan}')
print(f'{cikarilan} çıkarıldıktan sonra: {meyveler}')
````

### 8.5 — `.sort()` → Sırala

````python
# Sayıları sıralama
notlar = [53, 81, 67, 96, 74]
notlar.sort() # default reverse=False olarak gelir
print(f"notlar'ın küçükten-büyüğe sıralaması: {notlar}")
````

````python
notlar.sort(reverse=True)
print(f"notlar'ın büyükten-küçüğe sıralaması: {notlar}")
````

````python
# Alfabetik sıralama
isimler = ['Ali', 'Mehmet', 'Fatma', 'Ayşe', 'Zeynep', 'Ece', 'Demir']
isimler.sort()
print(f"isimler'in alfabetik olarak sıralaması: {isimler}")
````

### 8.6 — `.reverse()` → Ters Çevir

````python
sayilar = [1, 2, 3, 4, 5]
sayilar.reverse()
print(sayilar)
````

### 8.7 — `.count()` → Kaç Kez Geçiyor?

````python
notlar = [85, 92, 78, 85, 100, 85, 67]
print(f"85 sayısı notlar listesinin içerisinde {notlar.count(85)} kez geçiyor.")
print(f"100 sayısı notlar listesinin içerisinde {notlar.count(100)} kez geçiyor.")
print(f"50 sayısı notlar listesinin içerisinde {notlar.count(50)} kez geçiyor.")
````

### 8.8 — `.index()` → İlk Bulunduğu İndeks

````python
meyveler = ['elma', 'armut', 'muz', 'çilek', 'armut']
print(f"'armut' kelimesi meyveler listesinde ilk olarak {meyveler.index('armut')}.indekste bulundu.")
print(f"'muz' kelimesi meyveler listesinde ilk olarak {meyveler.index('muz')}.indekste bulundu.")
print(f"'ananas' kelimesi meyveler listesinde ilk olarak {meyveler.index('ananas')}.indekste bulundu.")
````

### 8.9 — `.clear()` → Tüm Elemanları Sil

````python
liste = [1, 2, 3, 4, 5]
liste.clear()
print(liste)
````

## Bölüm 9 — `in` Operatörü: Listede Var mı?

Bir elemanın listede olup olmadığını kontrol etmek için `in` kullanırız.  
Sonuç bir **boolean** döner: `True` veya `False`.

```python
eleman in liste   # True veya False
```

Bu, `if` ile birlikte çok güçlü bir araç olur!

````python
meyveler = ['elma', 'armut', 'muz', 'çilek']
print('elma' in meyveler) # true
print('karpuz' in meyveler) # false
````

````python
meyveler = ['elma', 'armut', 'muz', 'çilek']
aranan = input('Hangi meyveyi arıyorsun?')

if aranan in meyveler:
    print(f'{aranan} listede var!')
else:
    print(f'{aranan} listede yok!')
````

| Uzunluk | `len(liste)` | — | Hayır | Sayı |
| Var mı? | `x in liste` | — | Hayır | `True` / `False` |

### Örnek Çözüm

Kendin denedikten sonra aşağıdaki çözüme bak:

````python
# ========================================
# MİNİ UYGULAMA: Sınıf Not Defteri
# Örnek Çözüm
# ========================================

# 1. Listeleri oluştur
ogrenciler = ["Ali", "Ayşe", "Mehmet", "Zeynep", "Can"]
notlar = [85, 42, 78, 95, 60]

# 2. Tüm öğrencileri ve notlarını yazdır
print("=" * 30)
print("    SINIF NOT DEFTERİ")
print("=" * 30)
print(f"{ogrenciler[0]:10s} → {notlar[0]}")
print(f"{ogrenciler[1]:10s} → {notlar[1]}")
print(f"{ogrenciler[2]:10s} → {notlar[2]}")
print(f"{ogrenciler[3]:10s} → {notlar[3]}")
print(f"{ogrenciler[4]:10s} → {notlar[4]}")
print("-" * 30)

# 3. En yüksek not
en_yuksek = max(notlar)
en_yuksek_indeks = notlar.index(en_yuksek)
print(f"En yüksek not: {en_yuksek} ({ogrenciler[en_yuksek_indeks]})")

# 4. En düşük not
en_dusuk = min(notlar)
en_dusuk_indeks = notlar.index(en_dusuk)
print(f"En düşük not : {en_dusuk} ({ogrenciler[en_dusuk_indeks]})")

# 5. Kullanıcıdan öğrenci arama
print("\n" + "-" * 30)
aranan = input("Bir öğrenci adı gir: ")

if aranan in ogrenciler:
    indeks = ogrenciler.index(aranan)
    print(f"{aranan}'ın notu: {notlar[indeks]}")
    
    # Not kontrolü
    if notlar[indeks] >= 50:
        print("Durum: Geçti")
    else:
        print("Durum: Kaldı")
else:
    print(f"{aranan} bu sınıfta yok.")

print("=" * 30)
````
